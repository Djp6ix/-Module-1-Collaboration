Sort an Array of 0s, 1s, and 2s
We want to sort an array containing only 0s, 1s, and 2s without using the built-in sort function.  
We will use the Dutch National Flag algorithm for a one-pass solution with constant space.

```python
def sort_012(arr):
    low = 0
    mid = 0
    high = len(arr) - 1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:  # arr[mid] == 2
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1
    return arr
```

```python
arr1 = [0, 1, 2, 0, 1, 2]
arr2 = [0, 1, 1, 0, 1, 2, 1, 2, 0, 0, 0, 1]

print("Sorted arr1:", sort_012(arr1))
print("Sorted arr2:", sort_012(arr2))
```

Binary Search
Given a sorted array `arr[]` and an integer `k`, find the position (0-based index) of `k` using binary search.  
If `k` does not exist, return -1.  
If multiple occurrences are present, return the smallest index.

```python
def binary_search(arr, k):
    left = 0
    right = len(arr) - 1
    result = -1  # to store the first occurrence

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == k:
            result = mid
            right = mid - 1  # continue searching on the left for the smallest index
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid - 1

    return result
```

```python
# Test examples
arr1 = [1, 2, 3, 4, 5]
k1 = 4

arr2 = [11, 22, 33, 44, 55]
k2 = 445

arr3 = [1, 1, 1, 1, 2]
k3 = 1

print("Index of", k1, "in arr1:", binary_search(arr1, k1))
print("Index of", k2, "in arr2:", binary_search(arr2, k2))
print("Index of", k3, "in arr3:", binary_search(arr3, k3))
```
