13.1 Write the current date as a string to the text file today.txt. 

```python
from datetime import date

# Get today's date
today = date.today()

# Write to file
with open("today.txt", "w") as file:
    file.write(str(today))
```

13.2 Read the text file today.txt into the string today_string. 

```python
# Read the date string back from the file
with open("today.txt", "r") as file:
    today_string = file.read()

today_string
```

13.3 Phrase the date from today_string. 

```python
from datetime import datetime

# Convert the string back into a date object
parsed_date = datetime.strptime(today_string, "%Y-%m-%d").date()

parsed_date
```

15.1 Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit. 

```python
%%writefile multiprocessing_task.py
import multiprocessing
import time
import random
from datetime import datetime

def print_time():
    wait = random.random()
    time.sleep(wait)
    print(f"Process waited {wait:.2f} seconds. Current time: {datetime.now()}")

if __name__ == "__main__":
    processes = []
    for _ in range(3):
        p = multiprocessing.Process(target=print_time)
        processes.append(p)
        p.start()

    for p in processes:
        p.join()
```
