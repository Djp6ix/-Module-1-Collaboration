How many seconds are in an hour? 
Finding the number of seconds in an hour would mean to multiply the number of seconds in a minute (60) by the number of minutes in an hour (60).

```
# Calculate seconds in an hour
60 * 60
```

3.2 Assign the result from 3.1 to a variable called 'seconds_per_hour'

```
seconds_per_hour = 60 * 60
seconds_per_hour
```

3.3 How many seconds are in a day? 
Use the variable 'seconds_per_hour' and multiply by 24 

```
seconds_per_hour * 24
```

3.4 Calculate seconds per day again, but this time save the result in a variable called 'seconds_per_day'. 

```
seconds_per_day = seconds_per_hour * 24 
seconds_per_day
```

3.5 Divide seconds_per_day by seconds_per_hour using floating-point (/) division. 

```
seconds_per_day / seconds_per_hour
```

3.6 Divide seconds_per_day divided by seconds_per_hour integer (//) division. 
Did this number agree with the floating-point value from the previous question, aside from the final .0? 

```
seconds_per_day // seconds_per_hour
```
